```python
# Initialize an empty list to represent the stack
stack = []

def add_stick(stick):
    """Add a stick to the top of the stack."""
    stack.append(stick)

def show_stack():
    """Display the stack from top to bottom."""
    print("Stack (Top to Bottom):")
    for stick in reversed(stack):
        print(stick)

# Adding sticks to the stack
add_stick("Stick 1")
add_stick("Stick 2")
add_stick("Stick 3")

# Displaying the stack
show_stack()
```

```python
from collections import deque
list_q = [1,2,3,4]
queue = deque()

# Enqueue
# queue.append(1)
# queue.append(2)
# queue.append(3)
# queue.append(4)

for i in list_q:
    queue.append(i)

print(queue)

# Dequeue
while queue:
    print(queue.popleft())
```

```python
from collections import deque
queue = deque()

for i in range(20):
    if i%3 == 0:
        print(i,'enqueue')
        queue.append(i)
        
    elif i%4 == 0:
        print(i,'dequeue')
        queue.popleft()

print(queue)
```

```python
def 열기(현재상자번호):
   if 현재상자번호 == 10:
       print("장난감을 찾았어요!")
   else:
       다음상자번호 = 현재상자번호 + 1
       print(f"{현재상자번호}번 상자를 열어요. 다음은 {다음상자번호}번 상자를 열어볼까요?")
       열기(다음상자번호)  # 여기서 재귀가 일어나요.

# 첫 번째 상자부터 시작해요.
열기(1)
```

```python
def open(num):
    if num == 10:
        print("장난감을 찾았어요!")
    else:
        next_num = num + 1
        print(f"{num}번 상자를 열어요. 다음은 {next_num}번 상자를 열어볼까요?")
        open(next_num)  # recursive call 재귀 호출
```
