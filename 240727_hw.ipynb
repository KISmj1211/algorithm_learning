문제: "마법 상점의 아이템 가격 정렬하기"

마법 상점에서 다양한 마법 아이템을 판매하고 있습니다. 마법 상점의 점원인 여러분은, 고객들이 원하는 마법 아이템을 쉽게 찾을 수 있도록 아이템을 가격 순서대로 정렬해야 합니다. 이 상점에는 다양한 마법 아이템이 있으며, 각 아이템의 가격은 마법 책에 적혀 있습니다. 병합 정렬(Merge Sort) 알고리즘을 사용하여 아이템 가격 리스트를 오름차순으로 정렬하세요.

입력:
- 마법 책에는 가격들이 담긴 리스트 arr이 적혀있습니다. (예: arr = [38, 27, 43, 3, 9, 82, 10])

출력:
- 마법 책을 참고해 오름차순으로 가격이 정렬된 리스트를 반환합니다. (예: [3, 9, 10, 27, 38, 43, 82])

제한 사항:
- 주어진 리스트의 길이는 1 이상 100 이하입니다.
- 리스트의 요소는 모두 정수입니다.

```python
def merge_sort(arr):
    if len(arr) < 2:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    return merge(left_sorted, right_sorted)

def merge(left, right):
    sorted_list = []
    left_index, right_index = 0, 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            sorted_list.append(left[left_index])
            left_index += 1
        else:
            sorted_list.append(right[right_index])
            right_index += 1

    sorted_list.extend(left[left_index:])
    sorted_list.extend(right[right_index:])
    
    return sorted_list

# 예시 아이템 가격 리스트
arr = [38, 27, 43, 3, 9, 82, 10]

# 정렬 결과 출력
sorted_prices = merge_sort(arr)
print("정렬된 가격 리스트:", sorted_prices)
```
