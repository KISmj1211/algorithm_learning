문제: 간단한 과일 수 계산기  
여러 종류의 과일을 세는 프로그램을 작성하려고 합니다. 각 과일의 종류는 "Apple", "Banana", "Orange", "Grape", "Pineapple", "Strawberry"로 주어집니다. 주어진 과일 리스트에서 각 과일의 개수를 세어서 반환하는 solution 함수를 작성하세요.  

매개변수 설명  
과일 종류가 들어있는 리스트 fruit_list가 solution 함수의 매개변수로 주어집니다.  

fruit_list의 원소는 각 과일의 종류를 나타내며 "Apple", "Banana", "Orange", "Grape", "Pineapple", "Strawberry" 중 하나입니다.
fruit_list의 길이는 1 이상 100 이하입니다.  
반환 값 설명  
각 과일의 개수를 세어서 리스트 형태로 반환합니다. 반환하는 리스트는 다음 순서를 따라야 합니다: ["Apple", "Banana", "Orange", "Grape", "Pineapple", "Strawberry"].  

예시  
fruit_list	return  
["Apple", "Banana", "Orange", "Apple", "Grape"]	[2, 1, 1, 1, 0, 0]  
예시 설명  
예시 #1  

"Apple"이 2개, "Banana"가 1개, "Orange"가 1개, "Grape"가 1개입니다.
"Pineapple"과 "Strawberry"는 없습니다.  
따라서 [2, 1, 1, 1, 0, 0]을 반환합니다.  

```python
def solution(fruit_list):
    # 각 과일의 개수를 세기 위한 딕셔너리 초기화
    fruit_count = {
        "Apple": 0,
        "Banana": 0,
        "Orange": 0,
        "Grape": 0,
        "Pineapple": 0,
        "Strawberry": 0
    }
    
    # 주어진 과일 리스트에서 각 과일의 개수를 셈
    for fruit in fruit_list:
        if fruit in fruit_count:
            fruit_count[fruit] += 1
    
    # 결과를 리스트 형태로 반환
    # return [fruit_count["Apple"],fruit_count["Banana"],
    #     fruit_count["Orange"],fruit_count["Grape"],
    #     fruit_count["Pineapple"],fruit_count["Strawberry"]
    # ]
    
    return list(fruit_count.values())

# 테스트
fruit_list = ["Apple", "Banana", "Orange", "Apple", "Grape"]
print(solution(fruit_list))  # 출력: [2, 1, 1, 1, 0, 0]

fruit_list = ["Apple", "Apple", "Apple", "Strawberry", "Strawberry"]
print(solution(fruit_list))  # 출력: [3, 0, 0, 0, 0, 2]



# 테스트
print(solution(["Apple", "Banana", "Orange", "Apple", "Grape"]))  # 출력: [2, 1, 1, 1, 0, 0]
print(solution(["Strawberry", "Strawberry", "Strawberry"]))  # 출력: [0, 0, 0, 0, 0, 3]
print(solution(["Apple", "Banana", "Pineapple"]))  # 출력: [1, 1, 0, 0, 1, 0]
```

1-1. 문제: 간단한 숫자 범위별 개수 계산기
여러 숫자가 주어졌을 때, 각 숫자가 어느 범위에 속하는지 세는 프로그램을 작성하려고 합니다. 숫자 범위는 다음과 같이 정의됩니다:

0 이상 10 미만
10 이상 20 미만
20 이상 30 미만
30 이상 40 미만
40 이상 50 미만
50 이상
주어진 숫자 리스트에서 각 숫자가 속하는 범위별 개수를 세어서 반환하는 solution 함수를 작성하세요.

매개변수 설명
숫자가 들어있는 리스트 numbers가 solution 함수의 매개변수로 주어집니다.
numbers의 원소는 0 이상 100 이하의 정수입니다.
numbers의 길이는 1 이상 100 이하입니다.

반환 값 설명
각 범위별로 숫자의 개수를 세어서 리스트 형태로 반환합니다. 반환하는 리스트는 다음 순서를 따라야 합니다: [0 이상 10 미만, 10 이상 20 미만, 20 이상 30 미만, 30 이상 40 미만, 40 이상 50 미만, 50 이상].

예시
numbers	                    return
[3, 7, 15, 23, 35, 45, 55]	[2, 1, 1, 1, 1, 1]
[50, 60, 70, 80, 90, 100]	[0, 0, 0, 0, 0, 6]
[1, 11, 21, 31, 41, 51]	    [1, 1, 1, 1, 1, 1]
예시 설명
예시 #1

0 이상 10 미만: 3, 7 → 2개
10 이상 20 미만: 15 → 1개
20 이상 30 미만: 23 → 1개
30 이상 40 미만: 35 → 1개
40 이상 50 미만: 45 → 1개
50 이상: 55 → 1개
따라서 [2, 1, 1, 1, 1, 1]을 반환합니다.

```python
def solution(numbers):
    range_counter = [0 for _ in range(6)]
    for number in numbers:
        if 0 <= number < 10:
            range_counter[0] += 1
        elif 10 <= number < 20:
            range_counter[1] += 1
        elif 20 <= number < 30:
            range_counter[2] += 1
        elif 30 <= number < 40:
            range_counter[3] += 1
        elif 40 <= number < 50:
            range_counter[4] += 1
        elif number >= 50:
            range_counter[5] += 1
    return range_counter

# 테스트
print(solution([3, 7, 15, 23, 35, 45, 55]))  # 출력: [2, 1, 1, 1, 1, 1]
print(solution([50, 60, 70, 80, 90, 100]))   # 출력: [0, 0, 0, 0, 0, 6]
print(solution([1, 11, 21, 31, 41, 51]))     # 출력: [1, 1, 1, 1, 1, 1]
```

문제: 과일 개수 계산기 (딕셔너리 반환)
여러 종류의 과일이 주어졌을 때, 각 과일의 개수를 세는 프로그램을 작성하려고 합니다. 과일 종류는 "Apple", "Banana", "Orange", "Grape", "Pineapple", "Strawberry"로 주어집니다. 주어진 과일 리스트에서 각 과일의 개수를 세어서 딕셔너리 형태로 반환하는 solution 함수를 작성하세요.

매개변수 설명
과일 종류가 들어있는 리스트 fruit_list가 solution 함수의 매개변수로 주어집니다.

fruit_list의 원소는 각 과일의 종류를 나타내며 "Apple", "Banana", "Orange", "Grape", "Pineapple", "Strawberry" 중 하나입니다.
fruit_list의 길이는 1 이상 100 이하입니다.
반환 값 설명
각 과일의 개수를 딕셔너리 형태로 반환합니다. 반환하는 딕셔너리의 키는 과일 이름이며, 값은 해당 과일의 개수입니다.

예시  
fruit_list	                                    return
["Apple", "Banana", "Orange", "Apple", "Grape"]	{"Apple": 2, "Banana": 1, "Orange": 1, "Grape": 1, "Pineapple": 0, "Strawberry": 0}
["Strawberry", "Strawberry", "Strawberry"]	{"Apple": 0, "Banana": 0, "Orange": 0, "Grape": 0, "Pineapple": 0, "Strawberry": 3}
["Apple", "Banana", "Pineapple"]	{"Apple": 1, "Banana": 1, "Orange": 0, "Grape": 0, "Pineapple": 1, "Strawberry": 0}
예시 설명  
예시 #1  
"Apple"이 2개, "Banana"가 1개, "Orange"가 1개, "Grape"가 1개입니다.
"Pineapple"과 "Strawberry"는 없습니다.  
따라서 {"Apple": 2, "Banana": 1, "Orange": 1, "Grape": 1, "Pineapple": 0, "Strawberry": 0}을 반환합니다.  

```python
def solution(fruit_list):
    fruit_counter = {"Apple": 0, "Banana": 0, "Orange": 0, "Pineapple": 0, "Grape": 0, "Strawberry": 0}
    for fruit in fruit_list:
        fruit_counter[fruit] += 1
    return fruit_counter
```
