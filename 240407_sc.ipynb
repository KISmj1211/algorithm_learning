```python
"""
1. 주어진 딕셔너리를 값에 따라 오름차순으로 정렬하는 프로그램을 작성하세요. 
2. 정렬된 결과를 새로운 딕셔너리로 반환하세요.
"""
scores = {'Frank':100, 'Alice': 95, 'Bob': 78, 'Charlie': 88, 'David': 92, 'Eva':75}

print(scores.items())

# 1. 점수순으로 정렬
def sorted_by_values(item):
    return item[1]

scores_value = sorted(scores.items(), key=sorted_by_values)
print(scores_value)

scores_value_lambda = sorted(scores.items(), key=lambda item: item[1])

# 2. 이름 순으로 사람을 정렬
def sorted_by_key(item):
    return item[0]

scores_key = sorted(scores.items(), key=sorted_by_key)
print(scores_key)

scores_key_lambda = sorted(scores.items(), key=lambda item: item[0])
```
