문제: "도서관 책 정리하기"

당신은 도서관 사서로 일하고 있습니다. 이번에 도서관에서는 대규모 정리를 하기로 했습니다. 수많은 책들이 헝클어져 있어서 이 책들을 정렬된 상태로 만들어야 합니다. 당신의 목표는 책들을 제목의 알파벳 순서대로 정렬하는 것입니다. 이 작업을 빠르고 효율적으로 하기 위해 병합 정렬(Merge Sort)을 사용하기로 했습니다.

1. 책 더미 준비하기
   도서관에는 여러 더미의 책들이 있습니다. 각각의 책 더미에는 책의 제목이 적혀 있습니다. 예를 들어, 다음과 같은 책 더미가 있다고 가정해 봅시다:
   책 더미: ["Harry Potter", "The Great Gatsby", "1984", "To Kill a Mockingbird", "Moby Dick"]

2. 최종 결과
   병합 정렬이 완료된 후, 책 더미는 제목의 알파벳 순서대로 정렬됩니다. 최종 정렬된 결과를 출력하세요.
책 더미: ["Harry Potter", "The Great Gatsby", "1984", "To Kill a Mockingbird", "Moby Dick"]
정렬된 책 더미: ["1984", "Harry Potter", "Moby Dick", "The Great Gatsby", "To Kill a Mockingbird"]


```python
books = ["Harry Potter", "The Great Gatsby", "1984", "To Kill a Mockingbird", "Moby Dick"]

def merge_sort(books):
    if len(books) < 2:
        return books
    
    
    mid = len(books) // 2
    left = books[:mid]
    right = books[mid:]
    
    # print('left', left)
    # print('right', right)
    
    return merge(merge_sort(left),merge_sort(right))
    

def merge(left, right):
    sorted_books = []
    
    i = 0 # left_index 
    j = 0 # right_index
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_books.append(left[i])
            i +=1
        else:
            sorted_books.append(right[j])
            j += 1
    
    sorted_books.extend(left[i:])
    sorted_books.extend(right[j:])

    return sorted_books

print(merge_sort(books))
```
