```python
def dfs_iterative(graph, start):
    visited = []  # 방문한 노드를 추적하는 리스트
    stack = [start]  # 시작 노드를 스택에 추가

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.append(node)
            # 인접 노드를 스택에 추가 (역순으로 추가하여 작은 번호부터 방문)
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    
    return visited


graph = {
    1: [2, 5, 9],
    2: [3],
    3: [4],
    4: [],
    5: [6, 8],
    6: [7],
    7: [],
    8: [],
    9: [10],
    10: []
}

# DFS 실행
start_node = 1
visited = dfs_iterative(graph, start_node)
print("DFS 방문 순서:", visited)
```

```python
def dfs_recursive(graph, node, visited):
    if node not in visited:
        visited.append(node)
        for neighbor in graph[node]:
            dfs_recursive(graph, neighbor, visited)

visited = []
start_node = 1
dfs_recursive(graph, start_node, visited)
print("DFS 방문 순서:", visited)
```

```python
def dfs_iterative_find_target(graph, start, target):
    visited = []  # 방문한 노드를 추적하는 리스트
    stack = [start]  # 시작 노드를 스택에 추가

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.append(node)
            if node == target:
                return visited  # 목표 노드를 찾으면 방문한 순서를 반환
            # 인접 노드를 스택에 추가 (역순으로 추가하여 작은 번호부터 방문)
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    
    return visited  # 목표 노드를 찾지 못한 경우, 방문한 순서를 반환
```

```python
# 예시 그래프 (인접 리스트 표현)
graph = {
    1: [2, 5, 9],
    2: [3],
    3: [4],
    4: [],
    5: [6, 8],
    6: [7],
    7: [],
    8: [],
    9: [10],
    10: []
}

# DFS 실행
start_node = 1
target_node = 6
visited_nodes = dfs_iterative_find_target(graph, start_node, target_node)
print("DFS 방문 순서:", visited_nodes)
```
