```python
import heapq

min_heap = [3,1,4,1,5,9,2,6,5,3,5]

heapq.heapify(min_heap)

print(min_heap)
```

```python
heapq.heappush(min_heap, 0)
print(min_heap)
```

```python
min_val = heapq.heappop(min_heap)

print(min_val)
print(min_heap)
```

```python
min_heap = [3,1,4,1,5,9,2,6,5,3,5]

max_heap = [-x for x in min_heap]

heapq.heapify(max_heap)
print(max_heap)
```

```python
heapq.heappush(max_heap, -7)
print(max_heap)
```
