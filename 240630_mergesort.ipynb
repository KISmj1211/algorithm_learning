## 전체 알고리즘 설명
- 분할(Divide):
배열을 중간을 기준으로 두 개의 하위 배열로 재귀적으로 분할합니다. 배열의 길이가 1 또는 0이 될 때까지 분할을 계속합니다.
- 정복(Conquer):
분할된 하위 배열을 재귀적으로 정렬합니다. 기본적으로 길이가 1 이하인 배열은 이미 정렬된 상태로 간주됩니다.
- 결합(Combine):
merge 함수를 사용하여 두 개의 정렬된 하위 배열을 병합합니다.


## Merge 함수 과정

left = [2, 4, 6]
right = [1, 3, 5]

### 병합 과정
- 초기 상태:
left = [2, 4, 6]
right = [1, 3, 5]
current = [None, None, None, None, None, None] (초기화된 상태)
i = 0, j = 0, k = 0
- 첫 번째 비교 및 병합:
left[0] (2) vs right[0] (1)
1 < 2이므로 current[0] = right[0] = 1
current = [1, None, None, None, None, None]
j = 1, k = 1 (인덱스 업데이트)

- 두 번째 비교 및 병합:
left[0] (2) vs right[1] (3)
2 < 3이므로 current[1] = left[0] = 2
current = [1, 2, None, None, None, None]
i = 1, k = 2 (인덱스 업데이트)

- 세 번째 비교 및 병합:
left[1] (4) vs right[1] (3)
3 < 4이므로 current[2] = right[1] = 3
current = [1, 2, 3, None, None, None]
j = 2, k = 3 (인덱스 업데이트)

- 네 번째 비교 및 병합:
left[1] (4) vs right[2] (5)
4 < 5이므로 current[3] = left[1] = 4
current = [1, 2, 3, 4, None, None]
i = 2, k = 4 (인덱스 업데이트)

- 다섯 번째 비교 및 병합:
left[2] (6) vs right[2] (5)
5 < 6이므로 current[4] = right[2] = 5
current = [1, 2, 3, 4, 5, None]
j = 3, k = 5 (인덱스 업데이트)

- 남은 요소 병합:
right 배열은 모두 병합되었으므로, left 배열의 남은 요소를 current에 추가합니다.
current[5] = left[2] = 6
current = [1, 2, 3, 4, 5, 6]
i = 3, k = 6 (인덱스 업데이트)

- 최종 상태
current = [1, 2, 3, 4, 5, 6]



### 코드 설명
#### Merge_sort
merge_sort 함수는 주어진 리스트 current를 입력으로 받아 병합 정렬을 수행합니다.
리스트의 길이가 2보다 작으면 아무 작업도 수행하지 않고 반환합니다.
리스트를 중간 지점으로 나누어 left와 right 리스트로 분할합니다.
merge_sort 함수를 재귀적으로 호출하여 left와 right 리스트를 정렬합니다.
merge 함수를 호출하여 두 정렬된 리스트를 병합합니다.
#### Merge
merge 함수는 current, left, right 세 리스트를 입력으로 받아 current 리스트를 left와 right를 병합하여 정렬된 상태로 만듭니다.
i, j, k는 각각 left, right, current 리스트의 인덱스를 가리킵니다.
left와 right 리스트의 요소를 비교하여 작은 요소를 current에 채웁니다.
남은 요소들을 current에 채웁니다.

```python
def merge(current, left, right):
    i = j = k = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            current[k] = left[i]
            i += 1
        else:
            current[k] = right[j]
            j += 1
        k += 1
    
    while i < len(left):
        current[k] = left[i]
        i += 1
        k += 1
    
    while j < len(right):
        current[k] = right[j]
        j += 1
        k += 1


def merge_sort(current):
    if len(current) < 2:
        return
    
    mid = len(current) // 2
    left = current[:mid]
    right = current[mid:]
    
    merge_sort(left)
    merge_sort(right)
    
    merge(current, left, right)



# 예시 사용
array = [5, 3, 8, 4, 2, 7, 1, 10]
merge_sort(array)
print(array)
```

문제 설명:
가장 저렴한 물건순대로 배열하려고 합니다. 주어진 가격 리스트를 병합 정렬(Merge Sort) 알고리즘을 사용하여 오름차순으로 정렬하세요.

입력
- 가격들이 담긴 리스트 `arr`가 주어집니다. (예: `arr = [38, 27, 43, 3, 9, 82, 10]`)

출력
- 오름차순으로 가격이 정렬된 리스트를 반환합니다. (예: `[3, 9, 10, 27, 38, 43, 82]`)

힌트
병합 정렬은 다음과 같은 단계로 이루어집니다:
1. 리스트를 절반으로 나눕니다.
2. 각 절반을 재귀적으로 병합 정렬합니다.
3. 두 정렬된 절반을 병합하여 전체 리스트를 정렬합니다.

제한 사항
- 주어진 리스트의 길이는 1 이상 100 이하입니다.
- 리스트의 요소는 모두 정수입니다.



```python
def merge_sort(arr):
    if len(arr) < 2:
        return arr

    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    return merge(merge_sort(left), merge_sort(right))

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

# 테스트 케이스
arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = merge_sort(arr)
print(sorted_arr)  # 출력: [3, 9, 10, 27, 38, 43, 82]
```
