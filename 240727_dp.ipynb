문제: "최대 점수 만들기 게임“

문제 설명:
여러분은 게임에서 최대점수를 도전을 하고 있습니다. 각 돌다리에는 여러 단계가 있고, 점수가 있어요.
여러분은 돌다리를 건널 때 한 번에 한 계단 또는 두 계단씩 걸을 수 있습니다.
목표는 가능한 많은 점수를 얻으면서 도착지에 도달하는 것입니다.
계단의 각 단계별 점수가 주어졌을 때, 도달할 수 있는 최대 점수를 구하세요.

예시:
계단 점수: [1, -5, 1, 3,- 7, 5, 2]


```python
def max_score(stairs):
    n = len(stairs)
    if n == 0:
        return 0
    if n == 1:
        return stairs[0]
    
    # dp[i]는 i번째 계단까지의 최대 점수
    dp = [0] * n
    dp[0] = stairs[0]
    dp[1] = max(stairs[0] + stairs[1], stairs[1])

    for i in range(2, n):
        dp[i] = max(dp[i - 1] + stairs[i], dp[i - 2] + stairs[i])

    return dp[-1]

# 예시 입력
stairs = [1, -5, 1, 3, -7, 5, 2]
print(max_score(stairs))
```

"사탕 가게“

문제 설명:
사탕 가게에 다양한 종류의 사탕이 있고, 각 사탕은 다른 양의 행복 점수를 줍니다.
당신은 주어진 양의 돈으로 최대한의 행복 점수를 얻으려고 합니다.
하나의 사탕을 여러 번 구입할 수 있습니다.
각 사탕의 가격과 행복 점수가 주어졌을 때, 사용할 수 있는 돈으로 얻을 수 있는 최대 행복 점수를 찾으세요.

예시:
사탕 정보 (가격, 행복 점수): [(1, 2), (3, 4), (5, 7)]
사용할 수 있는 돈: 6

```python
def max_happiness(candies, money):
    dp = [0] * (money + 1)
    # print(dp)
    
    for cost, happiness in candies:
        # print(cost, happiness)
        for i in range(cost, money + 1):
            dp[i] = max(dp[i], dp[i - cost] + happiness)
            # print(dp)
    return dp[money]

candies_info = [(1, 2), (3, 4), (5, 7)]  # 사탕가격,행복점수
available_money = 6   # 사용할 수 있는 돈

max_happiness_score = max_happiness(candies_info, available_money)
print(max_happiness_score)
```
