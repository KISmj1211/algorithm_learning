
당신은 환상적인 식물 정원사입니다. 당신의 정원에는 아주 특별한 식물이 있습니다. 이 식물은 특정한 패턴에 따라 자랍니다. 첫 번째 해에는 1개의 잎이 생기고, 두 번째 해에는 1개의 잎이 더 생깁니다. 그 다음 해부터는 이전 두 해의 잎의 수를 더한 만큼의 잎이 생깁니다.
예를 들어:
- 첫 번째 해: 1개의 잎
- 두 번째 해: 1개의 잎
- 세 번째 해: 2개의 잎 (1 + 1)
- 네 번째 해: 3개의 잎 (1 + 2)
- 다섯 번째 해: 5개의 잎 (2 + 3)
주어진 해의 수(years) 동안 식물이 얼마나 많은 잎을 가지게 되는지를 재귀적으로 계산하는 함수를 작성하세요.
# 테스트 예시
print(calculate_leaves(1))  # 출력: 1
print(calculate_leaves(2))  # 출력: 1
print(calculate_leaves(3))  # 출력: 2
print(calculate_leaves(4))  # 출력: 3

```python
def calculate_leaves(years):
    if years <= 0:
        return 0
    if years == 1 or years == 2:
        return 1
    else:
        return calculate_leaves(years - 1) + calculate_leaves(years - 2)
    
```
