|-------------|-----------|-----------|---------|--------|
| 6           | 20        | 7         | 15      | 25     |

### 문제: 특정 날짜까지 남은 D-day 계산기

**배경**: 한 회사에서 특정 이벤트까지 남은 날짜를 계산하는 프로그램을 작성하려고 합니다. 기준 날짜와 목표 날짜가 주어질 때, 목표 날짜까지 며칠 남았는지(D-day)를 계산하세요.

이를 위해 다음과 같은 단계를 수행합니다:
1. 기준 날짜와 목표 날짜를 입력받습니다.
2. 기준 날짜로부터 목표 날짜까지의 일수를 계산합니다.

### 매개변수 설명

- 기준 날짜의 연, 월, 일을 나타내는 `current_year`, `current_month`, `current_day`와 목표 날짜의 연, 월, 일을 나타내는 `target_year`, `target_month`, `target_day`가 `solution` 함수의 매개변수로 주어집니다.
- 모든 날짜는 유효한 날짜입니다.

### 반환 값 설명

- 기준 날짜로부터 목표 날짜까지 남은 일수를 반환합니다.

### 예시

| current_year | current_month | current_day | target_year | target_month | target_day | return |
|--------------|---------------|-------------|-------------|--------------|------------|--------|
| 2024         | 5             | 26          | 2025        | 12           | 25         | 578    |

##### 예시 설명
- 기준 날짜는 2024년 5월 26일이고, 목표 날짜는 2025년 12월 25일입니다.
- 기준 날짜로부터 목표 날짜까지 578일 남았습니다.

### 설명

1. **days_in_month 함수**:
   - 주어진 월에 대해 해당 월의 일수를 반환합니다.

2. **days_since_start_of_year 함수**:
   - 주어진 월과 일에 대해 해당 연도에서 그 날짜까지의 일수를 계산합니다.

3. **days_in_year 함수**:
   - 윤년을 고려하지 않고 연도의 일수를 반환합니다.

4. **solution 함수**:
   - 기준 날짜와 목표 날짜가 같은 연도에 있는 경우, 두 날짜의 차이를 계산하여 반환합니다.
   - 기준 날짜와 목표 날짜가 다른 연도에 있는 경우:
     - 현재 연도의 남은 일수를 계산합니다.
     - 중간 연도들의 일수를 계산합니다.
     - 목표 연도의 경과한 일수를 계산합니다.
     - 세 부분을 합산하여 총 일수를 반환합니다.

```python
def days_in_month(month):
    month_list = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    return month_list[month - 1]

def days_since_start_of_year(month, day):
    days = day
    for m in range(1, month):
        days += days_in_month(m)
    return days

def days_in_year():
    return 365  # 윤년 고려하지 않음

def solution(current_year, current_month, current_day, target_year, target_month, target_day):
    current_day_of_year = days_since_start_of_year(current_month, current_day)
    target_day_of_year = days_since_start_of_year(target_month, target_day)
    
    if current_year == target_year:
        return target_day_of_year - current_day_of_year
    
    # current_year의 남은 일수 계산
    days_left_in_current_year = days_in_year() - current_day_of_year
    
    # 중간 연도들의 일수 계산
    days_in_between_years = (target_year - current_year - 1) * days_in_year()
    
    # target_year의 지난 일수 계산
    return days_left_in_current_year + days_in_between_years + target_day_of_year

# 테스트 케이스
current_year = 2024
current_month = 5
current_day = 26
target_year = 2025
target_month = 12
target_day = 25
ret = solution(current_year, current_month, current_day, target_year, target_month, target_day)

print("Solution: return value of the function is", ret, ".")
```

숫자 배열에서 최대값 찾기  
문제 설명:  
주어진 숫자 배열에서 최대값을 찾아 반환하는 프로그램을 작성하세요.  
입력: numbers: 숫자 배열.  
출력: 배열 내 최대값.  

```python
def find_maximum(numbers):
    max_num = 0 # 최댓값을 저장할 변수를 선언
    # for loop 사용하기
    for number in numbers:
        if number > max_num:
            max_num = number
    return max_num
```
